```python
from transformers import pipeline

classifier = pipeline("text-classification", model="unitary/toxic-bert")


text ="You are amazing "

result = classifier(text)

print(result)
```

```python
texts = [
    "You're a genius and I admire your work.",
    "Go die, nobody likes you.",
    "What a wonderful presentation!",
    "You stupid moron, get lost."
]

for t in texts:
    result = classifier(t)
    print(f"Text: {t}")
    print(f"Result: {result}\n")

```

```python
user_text = input("Enter any text to test for toxicity: ")
result = classifier(user_text)
print(f"Text: {user_text}")
print(f"Prediction: {result}")
```

```python
from sentence_transformers import SentenceTransformer, util

model = SentenceTransformer('paraphrase-MiniLM-L6-v2')

sent1 = "The quick brown fox jumps over the lazy dog."
sent2 = "A fast brown fox leaps above a sleepy dog."

embedding1 = model.encode(sent1, convert_to_tensor=True)
embedding2 = model.encode(sent2, convert_to_tensor=True)

similarity = util.pytorch_cos_sim(embedding1, embedding2)

print(f"Similarity Score: {similarity.item()}")
```
